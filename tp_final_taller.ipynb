```python
ips = ['192.168.1.5', '203.0.113.8', '203.0.113.5', '198.51.100.7', '192.168.1.3', '192.168.1.1', '192.168.1.4', '192.168.1.2']
counts = [121, 113, 129, 139, 116, 135, 133, 114]

fig, ax = plt.subplots()

bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:cyan']

ax.bar(ips, counts, color=bar_colors[:len(ips)])  # Ensure colors match the number of IPs

ax.set_ylabel('Access Counts')
ax.set_xlabel('IP Adresses')
ax.set_title('Access Counts by IP Address')

plt.xticks(rotation=45)

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

códigos = ['200', '301', '502', '404', '403', '500']
conteo = [401, 126, 112, 119, 120, 122]

fig, ax = plt.subplots()
ax.pie(conteo, labels=códigos, autopct='%1.1f%%', startangle=90)

ax.axis('equal')

plt.title('Proporción de Códigos HTTP')
plt.show()
```

El gráfico circular presenta la proporción de diferentes códigos HTTP generados por el servidor durante un período de tiempo determinado. Cada segmento del gráfico representa un código de estado específico y su porcentaje relativo con respecto al total de accesos registrados.


El gráfico de barras presenta el número de accesos al sitio web por cada dirección IP que ha realizado solicitudes al servidor durante un período de tiempo específico. Cada barra representa una dirección IP y su altura indica la cantidad de veces que esa IP accedió al sitio.
