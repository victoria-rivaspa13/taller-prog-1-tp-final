```python
ips = ['192.168.1.5', '203.0.113.8', '203.0.113.5', '198.51.100.7', '192.168.1.3', '192.168.1.1', '192.168.1.4', '192.168.1.2']
counts = [121, 113, 129, 139, 116, 135, 133, 114]

fig, ax = plt.subplots()

bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:cyan']

ax.bar(ips, counts, color=bar_colors[:len(ips)])  # Ensure colors match the number of IPs

ax.set_ylabel('Access Counts')
ax.set_xlabel('IP Adresses')
ax.set_title('Access Counts by IP Address')

plt.xticks(rotation=45)

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

cdigos = ['200', '301', '502', '404', '403', '500']
conteo = [401, 126, 112, 119, 120, 122]

fig, ax = plt.subplots()
ax.pie(conteo, labels=cdigos, autopct='%1.1f%%', startangle=90)

ax.axis('equal')

plt.title('Proporcin de Cdigos HTTP')
plt.show()
```
